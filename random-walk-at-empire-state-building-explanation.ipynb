## Capstone problem: Random walk at Empire state building

You are walking up the Empire State Building and you are playing a game with a friend. You throw a die 100 times.

Conditions=>

1. If its 1 or 2, you go one step down.
2. If its 3, 4 or 5, you go one step up.
3. If its 6, you throw the die again and walk the resulting number of steps.
4. Offcourse you can't go bellow step 0.
5. There is a 0.1% chance of you falling down the stairs. Then you will stat over from step=0

You bet your friend that you will reach 60 steps or higher. What are the chances, that you win this bet?

### step 1: Import necessary libraries.

### Step 2: Initialize pseudo random output with seed() function

### Step 3: Start the step counter from 0

```python
all_walks=[]          # starting the counter at '0'. That is why, we are starting with a blank square bracket
```

### Step 4: initiate for loop with range() function to run the simulation for 500 times.

Python range() Function

#### Definition and Usage:

The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

#### Syntax

range(start, stop, step)

* start => Optional. An integer number specifying at which position to start. Default is 0
* stop => Required. An integer number specifying at which position to stop (not included).
* step => Optional. An integer number specifying the incrementation. Default is 1

#### Helpful links for understanding if,elif,else functions :

#### Helpful links for understanding append() function :

```python
for i in range(500):    # 500 because we are trying to climb the building 500 times
    random_walk=[0]     # random_walk, counts number of steps for each of the 500 times we will walk up the building
    for x in range(100):       # value 100 symbolizes 100 dice throws as per the conditions
        step=random_walk[-1]      # fetching the last value. Since this is the last step count after calculation. The last step gives us the final number of steps that we were able to go up per trial
        dice=np.random.randint(1,7)  # rolling the dice and randomly getting the values from 1-6(7 is excluded here.It's upto 7). Because a die has 6 sides from 1-6
    
        if dice <=2:        # condition 1: If its 1 or 2, you go one step down.
            step=max(0,step-1)  # used max here to ensure that decrement can go max upto 0, since step cant be negative. Also used 'step-1' to ensure you go 1 step down since the dice rolled a 1 or 2 at this step
        elif dice <=5:     # condition 2: If its 3, 4 or 5, you go one step up.
            step=step+1    # 'Step+1' ensures you go 1 step up as the dice rolled anywhere between 3-5 as per the condition
        else:              
            step=step+np.random.randint(1,7) #condition 3:If its 6,you throw the die again & walk the resulting number of steps. For example you get 6 at first try and then 3 at second trial. You will go 3 steps then. 'step+np.random.randint(1,7)' ensures that task
        if np.random.rand() <= 0.001:  # Condition 5: There is a 0.1% chance of you falling down the stairs. Then you will stat over from step=0
            step=0                     # resetting from step=0 due to falling down (condition:5)
        
        random_walk.append(step)  # make sure to indent this within the ' for x in range(100):' loop. otherwise there will be errors
    all_walks.append(random_walk) # make sure to indent this outside the ' for x in range(100):' loop. otherwise there will be errors
```

### Step 5 : Transposing

I got confused at this point initially about why we were transposing the array. So it is important to understand this part.

The last line of code above appends the steps to 'all_walks'. If I run 'all_walks' here to see the output, then it will be hard to follow the note since the output is very lengthy. But to see how the output looks like for yourself, you can remove the hash (#) from the cell below and run 'all_walks'. This will give a sense of the structure.

```python
# all_walks
```

Now, the last step of every trial was at the end of each row in 'all_walks'. To get our desired output, we will transpose 'all_walks' so that the then rows are transposed to now columns. So in theory, last value of each column will be out steps reached per trail. We will have a total of 500 values since we tried for 500 time. 

This will look something like below=>

```python
np_aw_t=np.transpose(np.array(all_walks)) # transposing the array sets the last value of each walk in the last of the column. (our desired value. We are concerned with the last value of each trial since that gives the maximum steps of our trial)
np_aw_t
```

now we will select the last values of each column with the code below=>

```python
ends=np_aw_t[-1,:]   # select the last row of the transposed array. The last row gives the value of each walking end step
ends   # we will have 500 outputs. Some greater than 60, some might be equal to 60 and some less than 60
```

## Step 6: Calculating chances of going 60 steps or higher

Here I formatted the float output to percentage value. Took some help from the link below=>

https://www.w3resource.com/python-exercises/string/python-data-type-string-exercise-36.php

```python
a=np.count_nonzero(ends>=60)   # as the bet was to reach 60 steps or higher. That is why we used 'ends>=60'
t=500                          # total attempts
percentage = "{:.2%}".format(a/t)         # finding out the chances(%) by dividing successful attempts by total attempts
print(percentage)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

np.random.seed(123)   # providing a seed value ensures that we can produce same random outputs everytime. Python outputs are not entirely random, they are pseudo random outputs
all_walks=[]          # starting the counter at '0'. That is why, we are starting with a blank square bracket
for i in range(500):    # 500 because we are trying to climb the building 500 times
    random_walk=[0]     # random_walk, counts number of steps for each of the 500 times we will walk up the building
    for x in range(100):       # value 100 symbolizes 100 dice throws as per the conditions
        step=random_walk[-1]      # fetching the last value. Since this is the last step count after calculation. The last step gives us the final number of steps that we were able to go up per trial
        dice=np.random.randint(1,7)  # rolling the dice and randomly getting the values from 1-6(7 is excluded here.It's upto 7). Because a die has 6 sides from 1-6
    
        if dice <=2:        # condition 1: If its 1 or 2, you go one step down.
            step=max(0,step-1)  # used max here to ensure that decrement can go max upto 0, since step cant be negative. Also used 'step-1' to ensure you go 1 step down since the dice rolled a 1 or 2 at this step
        elif dice <=5:     # condition 2: If its 3, 4 or 5, you go one step up.
            step=step+1    # 'Step+1' ensures you go 1 step up as the dice rolled anywhere between 3-5 as per the condition
        else:              
            step=step+np.random.randint(1,7) #condition 3:If its 6,you throw the die again & walk the resulting number of steps. For example you get 6 at first try and then 3 at second trial. You will go 3 steps then. 'step+np.random.randint(1,7)' ensures that task
        if np.random.rand() <= 0.001:  # Condition 5: There is a 0.1% chance of you falling down the stairs. Then you will stat over from step=0
            step=0                     # resetting from step=0 due to falling down (condition:5)
        
        random_walk.append(step)  # make sure to indent this within the ' for x in range(100):' loop. otherwise there will be errors
    all_walks.append(random_walk) # make sure to indent this outside the ' for x in range(100):' loop. otherwise there will be errors
    
np_aw_t=np.transpose(np.array(all_walks)) # transposing the array sets the last value of each walk in the last of the column. (our desired value. We are concerned with the last value of each trial since that gives the maximum steps of our trial)
ends=np_aw_t[-1,:]   # select the last row of the transposed array. The last row gives the value of each walking end step

a=np.count_nonzero(ends>=60)   # as the bet was to reach 60 steps or higher. That is why we used 'ends>=60'
t=500                          # total attempts
percentage = "{:.2%}".format(a/t)         # finding out the chances(%) by dividing successful attempts by total attempts
print(percentage)
```
